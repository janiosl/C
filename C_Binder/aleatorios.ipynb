**Sorteio com números aleatórios**

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <unistd.h>

int main()
{
    char alunos[][80] = {"Janio", "Thiago", "Cristiane", "Andre"};
    int sorteados[4];
    int sorteado;
    int i;
    
    
    /*Listagem de alunos da turma*/
    puts("===========================================");
    puts("Alunos da turma:");
    puts("--------------------------------------------");
    
    for (i = 0; i < 4; i++)
    {
        printf("%i. %s\n", i+1, alunos[i]);
    }
    
    
    /*Semente aleatoria*/
    srand(time(NULL));
    
    /*Sorteio da ordem dos alunos*/
    puts("\nAlunos sorteados para apresentação:");
    puts("--------------------------------------------");
    
    for (i = 0; i < 4; i++)
    {
        /*Sorteio de numero aleatório entre 0 e 3*/
        sorteado = rand() % 3;
        sorteados[i] = sorteado;
    }
    
    /*Listagem do resultado do sorteio*/
    for (i = 0; i <2; i++)
        printf("%iº sorteado: %s\n", i+1, alunos[sorteados[i]]);
    puts("===========================================");
    
    return 0;
}
```

**Teste das funções srand e rand**

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main()
{
    int sorteado;
    
    srand(time(NULL));
    sorteado = rand() % 3;
    
    printf("Sorteado: %i", sorteado);
    
    return 0;
}
```

**Conferência do conteúdo de um vetor**

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main()
{
    int numbers[4];
    int i;
    int logica;   
    int n = 2;
    int pos;
    
    /*Define a posição aleatória para inserir o número para comparação*/
    srand(time(NULL));
    pos = rand() % 3;
    
    /*Insere o número de comparação no vetor de números*/
    numbers[pos] = n;
    
    /*Executa a comparação*/
    printf("Número para comparação: %i\n\n", n);
    
    for (i = 0; i < 4; i++)
    {
        if (n == numbers[i])
            logica = 1;
        else
            logica = 0;
            
                   
        printf("Número na posição %i: %i\n", i, numbers[i]);
        printf("Igual? %i\n\n", logica);
        
    }
    return 0;
}
```
